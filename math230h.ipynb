arclength $s=\int_a^b||\mathbf{r}'(t)||dt$

$dt$ is an infinitesimal piece of time

$ds$ an infinitesimal piece of arclength, the distance traveled in $dt$

infinitesimal distance traveled is equal to the product of speed and the infinitesimal change in time

$ds=||\mathbf{r}'(t)||dt=||\mathbf{v}(t)||dt$

units for $ds$ is distance (meters)

units for $||\mathbf{r}'(t)||$ is speed (meters per second)

units for $dt$ is time (seconds)

$s=\int ds$ the total is the sum of the parts/pieces

$\therefore s=\int ||\mathbf{r}'(t)||dt$
